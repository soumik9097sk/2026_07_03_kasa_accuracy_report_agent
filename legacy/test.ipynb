```python
import pandas as pd
import io
import json


# 2. Read the tab-separated data into a pandas DataFrame.
df = pd.read_csv(r"C:\Users\soumik.karmakar\Downloads\bquxjob_69a3c81c_19f37878830.csv")
df['month'] = df['month'].map({'JAN':1,
'FEB':2,
'MAR':3,
'APR':4,
'MAY':5,
'JUN':6,
'JUL':7,
'AUG':8,
'SEPT':9,
'OCT':10, 
'NOV':11,
'DEC':12})
# df['week_end_date'] = pd.to_datetime(df['week_end_date'])
# 3. Initialize the dictionary that will hold the final structure.
time_hierarchy = {}

# 4. Group the DataFrame by 'year' and then by 'month' to build the nested structure.
for year, year_df in df.groupby('year'):
    # Create a dictionary for the current year.
    time_hierarchy[year] = {}
    for month, month_df in year_df.groupby('month'):
        # For each month, convert the week details into a list of dictionaries.
        week_info = month_df[['week_end_date', 'weeknumber']].to_dict('records')
        # Assign the list of week details to the current month.
        time_hierarchy[year][month] = week_info

# 5. Print the final dictionary in a readable format.
print(json.dumps(time_hierarchy, indent=4))
```
